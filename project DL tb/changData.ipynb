```python
import pandas as pd
import numpy as np
```

```python
def get_dataset():
    print('Loading Dataset')
    #Importing Data from CSV file
    data=pd.read_csv("heart.csv")
#     labels=data.iloc[:,[0]].values
#     pixels=data['pixels']

#     #converting pixels to Gray Scale images of 48X48 
#     images = np.array([np.fromstring(pixel, dtype=int, sep=" ")for pixel in pixels])
#     #images=images/255.0 #normalize images between [0,1]
#     images = images.reshape(images.shape[0],48,48).astype('float32')
#     return images, labels
    return data
```

```python
data = get_dataset()

```

```python
data_copy = data.copy()
sex = set(data['Sex'])
chestPainType = set(data['ChestPainType'])
restingECG = set(data['RestingECG'])
exerciseAngina = set(data['ExerciseAngina'])
ST_Slope = set(data['ST_Slope'])
```

```python
print(sex)
print(chestPainType)
print(restingECG)
print(exerciseAngina)
print(ST_Slope)
```

```python
count = 0
for s in sex:
    data_copy['Sex']= data_copy['Sex'].replace([s], count)
    count +=1
    
count = 0
for c in chestPainType:
    data_copy['ChestPainType']= data_copy['ChestPainType'].replace([c], count)
    count +=1
    
count = 0
for r in restingECG:
    data_copy['RestingECG']= data_copy['RestingECG'].replace([r], count)
    count +=1
count = 0
for e in exerciseAngina:
    data_copy['ExerciseAngina']= data_copy['ExerciseAngina'].replace([e], count)
    count +=1
count = 0
for s in ST_Slope:
    data_copy['ST_Slope']= data_copy['ST_Slope'].replace([s], count)
    count +=1
data_copy
```

```python
num_list= ["Age","RestingBP","Cholesterol","FastingBS","MaxHR","Oldpeak", 'ChestPainType','RestingECG','ST_Slope']
for i in num_list:
    mean = data_copy[i].mean()
    std = data_copy[i].std()
    data_copy[i] = (data_copy[i]-mean)/std
data_copy
```

```python
data_np = data_copy.to_numpy()
```

```python
np.save('heart_convert2', data_np)
```
